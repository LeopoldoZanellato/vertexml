```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
import warnings
from datetime import datetime
import joblib

import os
import uuid

import google.cloud.aiplatform as aiplatform

# Ignora todos os avisos
warnings.filterwarnings('ignore')
```

```python
# --- 1. Definição do Caminho no GCS ---
GLOBAL_BASE_PATH = "gs://the_chustomer_churn/"
GCS_BASE_PATH = "gs://the_chustomer_churn/dataset/"
RAW_DATASET_NAME = "dataset.csv"
RAW_DATASET_PATH = f"{GCS_BASE_PATH}{RAW_DATASET_NAME}"
TARGET_COLUMN = 'Churn'
PROJECT_ID = 'vertexai-457414'
LOCATION = "us-central1" 
EXPERIMENT_BUCKET = "gs://the_chustomer_churn/dataset/experiments/"

variable_params = {}
variable_params['GCS_BASE_PATH'] = GCS_BASE_PATH
variable_params['RAW_DATASET_NAME'] = RAW_DATASET_NAME
variable_params['RAW_DATASET_PATH'] = RAW_DATASET_PATH
variable_params['TARGET_COLUMN'] = TARGET_COLUMN
variable_params['PROJECT_ID'] = PROJECT_ID
variable_params['LOCATION'] = LOCATION
variable_params['EXPERIMENT_BUCKET'] = EXPERIMENT_BUCKET
```

```python
# --- Configurações Explícitas ---

agora = datetime.now().strftime("%Y-%m-%d-%H-%M-%S")

EXPERIMENT_NAME = "telco-customer-v1"
# ---------------------------------

aiplatform.init(
    project=PROJECT_ID,
    location=LOCATION,
    staging_bucket=EXPERIMENT_BUCKET,
    experiment=EXPERIMENT_NAME,

    #experiment_tensorboard=False
)
RUN_NAME = f"run-{agora}"
variable_params['RUN_NAME'] = RUN_NAME
# Agora você pode começar sua run
aiplatform.start_run(RUN_NAME)
aiplatform.log_params(variable_params)
```

```python

# Proporções
TEST_SIZE = 0.15  # 15% para teste
VAL_SIZE = 0.15   # 15% para validação
# (Treino será os 70% restantes)

dataset_params = {}
dataset_params['TEST_SIZE'] = TEST_SIZE
dataset_params['VAL_SIZE'] = VAL_SIZE

aiplatform.log_params(dataset_params)
```

```python
# pip install fsspec gcsfs pandas seaborn scikit-learn matplotlib # pip install google-cloud-aiplatform
dataset = pd.read_csv(f"{RAW_DATASET_NAME}")
```

```python
# Primeiro, verificamos se a coluna-alvo existe
if TARGET_COLUMN not in dataset.columns:
    print(f"!!! ERRO: A coluna '{TARGET_COLUMN}' não foi encontrada no dataset.")
    print(f"Por favor, atualize a variável 'TARGET_COLUMN' com o nome correto.")
    print(f"Colunas disponíveis: {dataset.columns.tolist()}")
else:
    print(f"Iniciando divisão estratificada pela coluna: '{TARGET_COLUMN}'")
    
    # Etapa 1: Separar o Teste (15%) do restante (85%)
    # O 'stratify' garante que 'test_df' e 'train_val_df' tenham a mesma proporção da coluna-alvo
    train_val_df, test_df = train_test_split(
        dataset,
        test_size=TEST_SIZE,
        random_state=42,
        stratify=dataset[TARGET_COLUMN]
    )


    # Etapa 2: Separar o restante (85%) em Treino (70%) e Validação (15%)
    # Precisamos recalcular a proporção de validação em relação ao 'train_val_df'
    # val_size_ratio = 15% / 85%
    val_size_ratio = VAL_SIZE / (1 - TEST_SIZE)
    
    train_df, val_df = train_test_split(
        train_val_df,
        test_size=val_size_ratio,
        random_state=42,
        stratify=train_val_df[TARGET_COLUMN] # Estratificar novamente
    )

    # --- 3. Verificação das Proporções ---
    print("-" * 40)
    print("Verificação das proporções (value_counts normalizado):")
    print("\nProporção Original no Dataset Completo:")
    print(dataset[TARGET_COLUMN].value_counts(normalize=True))
    
    print(f"\nProporção no Treino ({len(train_df)} linhas):")
    print(train_df[TARGET_COLUMN].value_counts(normalize=True))
    
    print(f"\nProporção na Validação ({len(val_df)} linhas):")
    print(val_df[TARGET_COLUMN].value_counts(normalize=True))
    
    print(f"\nProporção no Teste ({len(test_df)} linhas):")
    print(test_df[TARGET_COLUMN].value_counts(normalize=True))
    print("-" * 40)


    # --- 4. Salvar os arquivos no Google Cloud Storage ---
    print(f"Salvando arquivos em {GCS_BASE_PATH}...")

    train_df.to_csv(f"{GCS_BASE_PATH}train.csv", index=False)
    val_df.to_csv(f"{GCS_BASE_PATH}validation.csv", index=False)
    test_df.to_csv(f"{GCS_BASE_PATH}test.csv", index=False)

    print("✅ Sucesso! Os 3 arquivos estratificados foram salvos no GCS.")
```

```python
import pandas as pd
import numpy as np

def preprocess_features(df_input):
    """
    Função de LIMPEZA: Corrige tipos, trata nulos e mapeia o alvo.
    Não faz 'fit' ou 'transform' (scaling ou encoding).
    """
    # 1. Copiar o dataframe
    df = df_input.copy()
    
    # --- ETAPA 1: Correção de Tipos e Nulos ---
    print(f"Limpando TotalCharges...")
    df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
    df['TotalCharges'].fillna(0, inplace=True)
    
    
    # --- ETAPA 2: Mapeamento APENAS DA COLUNA-ALVO ---
    # O 'y' (target) precisa ser uma coluna única (0 ou 1)
    if 'Churn' in df.columns:
        print(f"Mapeando coluna 'Churn' para 0/1...")
        df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

        
    # --- ETAPA 3: Preparação das Categóricas ---
    # Apenas garante que SeniorCitizen seja 'str' para o OneHotEncoder
    if 'SeniorCitizen' in df.columns:
        df['SeniorCitizen'] = df['SeniorCitizen'].astype(str)
        
    
    # --- ETAPA 4: Remoção de Colunas ---
    if 'customerID' in df.columns:
        df = df.drop(columns=['customerID'])
        
    return df
```

```python
# --- 1. Rodar a função de LIMPEZA ---
# (Assumindo que df_train_raw, df_val_raw, df_test_raw estão carregados)

print("Processando dataset de TREINO...")
df_train = preprocess_features(df_train_raw)

print("Processando dataset de VALIDAÇÃO...")
df_val = preprocess_features(df_val_raw)

print("Processando dataset de TESTE...")
df_test = preprocess_features(df_test_raw)

print("\n✅ Limpeza inicial concluída!")
```

```python
df_train
```

```python
# --- 2. Separar X (features) e y (alvo) ---
# O 'y' já está pronto (0/1) graças à nossa função.
# Usamos .pop() para extrair o 'y' e já removê-lo do 'X'

y_train = df_train.pop('Churn')
y_val = df_val.pop('Churn')
y_test = df_test.pop('Churn')

# O que sobra em df_train, df_val, df_test é o nosso X
X_train = df_train
X_val = df_val
X_test = df_test


# --- 3. Definir Colunas Numéricas e Categóricas (para o Pipeline) ---

numeric_features = ['tenure', 'MonthlyCharges', 'TotalCharges']

categorical_features = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents', 
    'PhoneService', 'MultipleLines', 'InternetService', 
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
    'TechSupport', 'StreamingTV', 'StreamingMovies', 
    'Contract', 'PaperlessBilling', 'PaymentMethod'
]

# --- 4. Criar o Preprocessor (O "Cérebro") ---

# Pipeline para features numéricas
numeric_transformer = MinMaxScaler()

# Pipeline para features categóricas
categorical_transformer = OneHotEncoder(
    drop='first',              
    handle_unknown='ignore',   # Ignora categorias raras que apareçam no teste
    sparse_output=False        
)

# Juntar tudo no ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='passthrough' 
)


# --- 5. FIT (Aprender) e TRANSFORM (Aplicar) ---

# FIT (APRENDER) - SOMENTE COM X_train
print("\n'Fitando' (aprendendo) o preprocessor com X_train...")
preprocessor.fit(X_train)

# TRANSFORM (APLICAR) - Em todos os datasets
print("Transformando X_train...")
X_train_processed = preprocessor.transform(X_train)

print("Transformando X_val...")
X_val_processed = preprocessor.transform(X_val)

print("Transformando X_test...")
X_test_processed = preprocessor.transform(X_test)

print("\n✅ Processamento com Pipeline concluído com sucesso!")
print(f"Formato do X_train processado: {X_train_processed.shape}")
print(f"Formato do X_val processado: {X_val_processed.shape}")
print(f"Formato do X_test processado: {X_test_processed.shape}")
```

```python
# y_train já deve estar como 0s e 1s da etapa anterior
# Conta quantos 'Não' (0) e quantos 'Sim' (1)
count_no = (y_train == 0).sum()
count_yes = (y_train == 1).sum()

# O peso é a proporção de Não / Sim
scale_pos_weight = count_no / count_yes

print(f"Dataset de treino:")
print(f"Não Churn (0): {count_no}")
print(f"Sim Churn (1): {count_yes}")
print(f"Proporção: 1 'Sim' para cada {scale_pos_weight:.2f} 'Não'")
print(f"Usaremos scale_pos_weight = {scale_pos_weight:.2f}")
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. Instanciar o modelo
# Usamos o 'scale_pos_weight' que acabamos de calcular
# 'random_state=42' garante que o modelo seja o mesmo toda vez que rodar
model_xgb = xgb.XGBClassifier(
    objective='binary:logistic',  # Classificação binária
    scale_pos_weight=20, 
    random_state=42,
    n_estimators=150,             # <<< MUDANÇA: número fixo de árvores
    eval_metric='logloss'         # (Mantido no construtor)
)
```

```python
# E no fit, você ainda precisa passar o eval_set
model_xgb.fit(
    X_train_processed, 
    y_train,
    eval_set=[(X_val_processed, y_val)],
    verbose=100
)
```

```python
fs = gcsfs.GCSFileSystem() # Inicia o sistema de arquivos do GCS

MODEL_PATH = f"{GLOBAL_BASE_PATH}run/{RUN_NAME}/artifacts/model.joblib"
# Abre o arquivo no GCS no modo "write bytes" ('wb')
with fs.open(MODEL_PATH, 'wb') as f:
    joblib.dump(model_xgb, f)

print("✅ Model salvo com sucesso no GCS (método explícito)!")


model_artifact = aiplatform.Artifact.create(
    schema_title="system.Model", 
    display_name="Model", 
    uri=MODEL_PATH,
)

model_param = {}

model_param['MODEL_NAME'] = str(model_xgb.__class__)
model_param['MODEL_PATH'] = MODEL_PATH
aiplatform.log_params(model_param)
```

```python
# 3. Fazer Previsões na VALIDAÇÃO
print("Avaliando o modelo no set de validação...")
y_preds_val = model_xgb.predict(X_val_processed)


# 4. Avaliar os Resultados
print("\n--- Relatório de Classificação (Validação) ---")
# 'target_names' deixa o relatório bonito com 'No' e 'Yes'
print(classification_report(y_val, y_preds_val, target_names=['No (0)', 'Yes (1)']))


# 5. Plotar a Matriz de Confusão
print("\n--- Matriz de Confusão (Validação) ---")
ConfusionMatrixDisplay.from_predictions(
    y_val, 
    y_preds_val, 
    display_labels=['No Churn', 'Yes Churn'],
    cmap='Blues'
)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("Extraindo a importância das features...")

# 1. Obter os nomes das features do preprocessor
#    Isso nos dá todos os nomes das colunas numéricas e do one-hot encoding
try:
    feature_names = preprocessor.get_feature_names_out()
except AttributeError:
    # Versões mais antigas do scikit-learn não têm get_feature_names_out()
    # Tentativa de workaround (pode ser menos limpo)
    print("Método get_feature_names_out() não encontrado, tentando workaround...")
    
    # Nomes numéricos
    num_names = numeric_features
    
    # Nomes categóricos
    # Acessa o encoder 'cat' treinado dentro do preprocessor
    cat_encoder = preprocessor.named_transformers_['cat']
    cat_names = cat_encoder.get_feature_names_out(categorical_features)
    
    # Junta todos os nomes na ordem correta
    feature_names = np.concatenate([num_names, cat_names])

# 2. Obter as pontuações de importância do modelo treinado
importances = model_xgb.feature_importances_

# 3. Criar um DataFrame para visualizar
df_importance = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
})

# 4. Ordenar o DataFrame pela importância
df_importance = df_importance.sort_values(by='Importance', ascending=False)

print("\n--- Top 10 Features Mais Importantes ---")
print(df_importance.head(10))


# --- 5. Plotar as Top 20 Features ---
print("\nPlotando o gráfico das 20 features mais importantes...")

plt.figure(figsize=(12, 10))
sns.barplot(
    x='Importance', 
    y='Feature', 
    data=df_importance.head(20),
    palette='viridis' # Você pode mudar a cor se quiser
)
plt.title('Top 20 Features Mais Importantes (XGBoost)', fontsize=16)
plt.xlabel('Importância (Gain)', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.show()
```

```python
import google.cloud.aiplatform as aiplatform
from sklearn.metrics import (
    classification_report, 
    ConfusionMatrixDisplay, 
    confusion_matrix, 
    roc_auc_score
)
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np # Importar numpy para o caso de precisar

# --- 3. Fazer Previsões (Probabilidades e Classes) ---
print("Avaliando o modelo no set de validação...")

y_preds_val = model_xgb.predict(X_val_processed)
y_probs_val = model_xgb.predict_proba(X_val_processed)[:, 1]


# --- 4. Calcular as Métricas ---

report_dict = classification_report(y_val, y_preds_val, output_dict=True)
auc_score = roc_auc_score(y_val, y_probs_val)
tn, fp, fn, tp = confusion_matrix(y_val, y_preds_val).ravel()


# --- 5. Criar o Dicionário de Métricas para o Vertex AI (COM A CORREÇÃO) ---

metrics_to_log = {
    # Métricas da classe '1' (Churn=Yes)
    "recall_yes": report_dict['1']['recall'],
    "precision_yes": report_dict['1']['precision'],
    "f1_score_yes": report_dict['1']['f1-score'],
    
    # Métricas gerais
    "accuracy": report_dict['accuracy'],
    "roc_auc": auc_score,
    
    # Números da Matriz de Confusão (CONVERTIDOS PARA INT)
    "true_positives": int(tp),      # <-- CORREÇÃO AQUI
    "false_positives": int(fp),     # <-- CORREÇÃO AQUI
    "true_negatives": int(tn),      # <-- CORREÇÃO AQUI
    "false_negatives": int(fn),     # <-- CORREÇÃO AQUI
}

# --- 6. SALVAR AS MÉTRICAS NO VERTEX AI EXPERIMENTS ---
print(f"Logando {len(metrics_to_log)} métricas na run atual...")
aiplatform.log_metrics(metrics_to_log) # Este comando agora vai funcionar
print("Métricas logadas com sucesso!")


# --- 7. Imprimir e Plotar (Para seu notebook) ---
print("\n--- Relatório de Classificação (Validação) ---")
print(classification_report(y_val, y_preds_val, target_names=['No (0)', 'Yes (1)']))

print(f"\nROC AUC Score: {auc_score:.4f}")

print("\n--- Matriz de Confusão (Validação) ---")
ConfusionMatrixDisplay.from_predictions(
    y_val, 
    y_preds_val, 
    display_labels=['No Churn', 'Yes Churn'],
    cmap='Blues'
)
plt.show()
```
